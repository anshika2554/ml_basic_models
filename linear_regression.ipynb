```python
import pandas as pd
import numpy as np
import sklearn
import matplotlib.pyplot as plt
```

##Creating Data


```python
x=10* np.random.rand(1000)
y=3*x+ np.random.randn(1000)
plt.scatter(x,y)
plt.show()
```

##Preprocessing

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x,y, test_size=0.3, random_state=0)
```

##Model Building and Prediction

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train.reshape(-1,1),y_train)
y_train_pred= model.predict(x_train.reshape(-1,1))
y_test_pred= model.predict(x_test.reshape(-1,1))
```

##Visualisation

```python
plt.figure(figsize=(8,6))
plt.scatter(y_train_pred, y_train_pred - y_train, color="b", marker="*", label="training data")
plt.scatter(y_test_pred, y_test_pred - y_test, color="orange", marker="o", label="testing data")
plt.xlabel("predicted values")
plt.ylabel("residual")
plt.hlines(y=0, xmin=-3, xmax=33, lw=2,color="k")
plt.xlim([-5,35])
plt.ylim([-25,15])
plt.legend()
plt.show()
```

##Evaluation

```python
#using mean_square error
from sklearn.metrics import mean_squared_error
mean_squared_error(y_train_pred, y_train)
mean_squared_error(y_test_pred, y_test)
```

```python
# by r2_score
from sklearn.metrics import r2_score
r2_score(y_train_pred, y_train)
r2_score(y_test_pred, y_test)
```
