```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
np.random.seed(42)
n_samples = 100
```

```python
x= np.linspace(0, 10, 100)
rng= np.random.randn(n_samples)*100
y = x**3 + rng +100
```

```python
plt.figure(figsize=(10,8))
plt.scatter(x,y)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
lr= LinearRegression()
lr.fit(x.reshape(-1,1),y)
model_pred = lr.predict(x.reshape(-1,1))
plt.scatter(x, y)
plt.plot(x, model_pred)
plt.show()
#it is not fitting well as line is linear
print(r2_score(y,model_pred))
```

# Polynomial

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_reg = PolynomialFeatures(degree=2)
x_poly = poly_reg.fit_transform(x.reshape(-1,1)) #suare the valus n x for cube put degree 3
```

```python
lr2 = LinearRegression()
lr2.fit(x_poly, y.reshape(-1,1))
y_pred = lr2.predict(x_poly)
```

```python
plt.scatter(x,y)
plt.plot(x,y_pred)
plt.show()
print(r2_score(y,y_pred))
```
